```python
# title 및 label 설정

plt.figure(figsize=(12,6))

x = np.linspace(0,3,101)
y = np.sqrt(x)

plt.plot(x,y)
plt.title("Graph of y = x^(0.5)", fontsize = 25)
plt.xlabel("x", fontsize = 20)
plt.ylabel("y", fontsize = 20)
plt.grid() # 그래프에 격자 표시
```

```python
# plot 스타일 조절 (색상, 선 종류 등등)
x = np.linspace(0, 2*np.pi, 101)
y = np.cos(x)

# plt.plot(x,y, color = 'red')
# plt.plot(x,y, color = 'g')
# plt.plot(x,y, color = 'red', linestyle = '--')
plt.plot(x,y,"b--") #파란색(b), 대시(--)라인으로 그려라
plt.show()

```

```python
# 여러개의 plot 가능

x = np.linspace(1,4,101)
y1 = 1/x
y2 = 1/x**2

plt.plot(x, y1,"r--", x, y2, "b-.") 
plt.legend(["1/x", "1/x^2"])
```

```python
# 보통 이렇게 한다.

x = np.linspace(1,4,101)
y1 = 1/x
y2 = 1/x**2

plt.plot(x,y1, 'r--', label = "1/x") # --
plt.plot(x,y2, 'b-.', label = "1/x^2") # -.
# 그래프 두 개를 따로 그렸다
# label을 통해 범례 이름을 지정해 주었다.
plt.legend(loc = 'best') # 'best': 범례를 자동으로 최적의 위치에 위치시킨다. 
```

```python
x = np.linspace(0,1,101)
y1 = x**2
y2 = x
y3 = np.sqrt(x)

plt.plot(x,y1,label="y=x^2")
plt.plot(x,y2,label="y=x")
plt.plot(x,y3,label="y=sqrt(x)")

plt.xlabel("x")
plt.ylabel("y")

plt.grid()
plt.legend( loc = "best")
plt.show()
```

```python
# seaborn : 통계적 그래프를 잘 보여주는 라이브러리

import seaborn as sns

tips = sns.load_dataset("tips")
# seaborn에서 사용할 수 있는 dataset 목록에는 titanic, tips, iris 등이 있다.
# sns.load_dataset('데이터셋 이름')
tips.head()
```

```python
# 데이터를 받으면 탐색을 먼저!

tips.describe()
```

```python
# 결측치의 여부
tips.isna().sum()
```

```python
# 상관관계 시각화 하기.

# correlation matrix 구하기
tips.corr()
```

```python
# heatmap으로 시각화 하기.

plt.figure(figsize = (8,8))
sns.heatmap(tips.corr(), annot = True, annot_kws = {"fontsize":19})
# sns.heatmap(~)
# annot: 각 cell의 값 표기 유무
# fmt: 각 cell의 값의 데이터 형태를 지정해줌
# annot_kws: 표시되는 각 cell 값들의 사이즈를 지정 
plt.show()
```

```python
# scatter plot : 산점도

# relplot 으로 시각화 한다.
# xlabel, ylabel 입력을 안해도 자동으로 뜬다.

# 총 가격이 높을수록 받는 팁도 많을까?
# -> total_bill과 tip은 어느정도로 상관관계가 있을까?
# => (당연히 다 그런것은 아니지만) 총 가격이 높을수록 tip도 많이 받는 편이다! 
# 총 가격이 높아질 수록 받는 팁도 상대적으로 높아지는 경향을 보인다 

sns.set(style = "darkgrid") # 스타일 설정. 앞으로 그리는 플롯 그림에 영향을 미친다.
sns.relplot(x = "total_bill", y = 'tip', data = tips) # relplot으로 시각화한다. 
# x = "total_bill" : x축 labelling
# y = "tip" : y축 labeling
# data 는 'tips' 데이터셋을 사용 
```

1. set(context, style, palette, font, font_scale, color_code, rc): set_theme과 동일한 함수이다. 스타일, 팔레트, 글꼴, 글꼴 크기 등을 설정. 이는 앞으로 그리게 되는 모든 플롯의 그림에 영향을 준다. 즉, 환경변수 설정이다. (추후 버전에서 삭제될 얘정이다)

2. set_theme(context, style, palette, font, font_scale, color_code, rc): 이 함수를 사용하면 앞으로 그려지는 plot들의 테마들이 일괄적으로 세팅된다. 이 함수를 사용하면 python.matploblib 내의 rcParams system을 변경하게되는데, 이렇게되면 전체적인 플롯들의 테마(글꼴, 팔레트 등)이 일률적으로 변경된다.

3. 
- scatterplot(): 산점도 그려줌
- lineplot(): 선 그래프 그려줌
- relplot(): 위 두 함수를 모두 포함하고 있다고 생각하면 된다.


```python
# 점심과 저녁간의 팁 차이가 있을까?
# -> hue를 이용하여 데이터를 구분하기.

sns.relplot(x = "total_bill", y = "tip",
            hue = 'time', height = 7, 
            data = tips)

# hue는 색을 의미. 데이터의 구분 기준을 정하여(여기서는 time) 색상을 통해 내용을 구분.
# => hue 인수에 카테고리 변수 이름을 지정하여 카테고리 값에 따라 색상을 다르게 할 수 있다.
```

```python
# style로 데이터 구분하기
# 점심과 저녁간의 팁중 성별에 따른 팁의 차이가 있을까?

sns.relplot(x = 'total_bill', y = 'tip',
            hue = 'time', style = 'sex', height = 6,
            data = tips)

# style이라는 조건을 통해 hue와 함께 이중으로 조건을 걸었다고 생각해라.
# hue가 색으로 구분하는 거라면, style은 표기를 다르게 해주는 것이다. 
# + size는 크기를 다르게 해서 구분해주는 것.
# hue, style에는 time, sex 등 카테고리 변수를 지정해준다. 

# 보기가 힘들다...
```

```python
# 데이터를 다른 기준에 따라 각각 볼 수 있다.
# -> col로 그림 나누기


sns.relplot(x = 'total_bill', y = 'tip',
            hue = 'time', col = 'sex', height = 6,
            data = tips)

```

=> 남자가 상대적으로 팁을 많이 내고, 더 높은 팁을 낸다.

```python
# Boxplot : 다섯 수치를 요약해주는 그래프 / 데이터의 분포를 잘 살펴볼 수 있다.  

# 1. 최솟값
# 2. 제 1사분위(Q1) : 데이터의 25% 위치
# 3. 제 2사분위(Q2) = Median(중앙값) : 50% 위치
# 4. 제 3사분위(Q3) : 데이터의 75% 위치
# 5. 최댓값

sns.set_theme(style = 'white')
sns.boxplot(x = "total_bill", data = tips)
plt.title("Total bill boxplot")
```

```python
# boxplot 여러개 그리기
# ex) 요일별 total_bill에 대한 boxplot

sns.boxplot(x = 'day', y = 'total_bill', data = tips)
# 요일별 total_bill에 대한 것이므로 x축에 'day', y축에 'total_bill
# x,y축을 이렇게 지정해주면 자동으로 여러개 그래프가 생성됨
```

```python
# 요일별 팁의 박스그래프

sns.boxplot(x="day", y="tip", data=tips)
```

```python
# 성별을 기준으로 팁에대한 박스그래프 그리기.

sns.boxplot(x = "sex", y = "tip", data = tips)
```

```python
# hue 넣기

sns.boxplot(x="day", y="tip", hue="sex", data=tips)

# => 성별에 따라 색으로 구분되어 그래프가 따로 나타난다. 
```

```python
titanic = sns.load_dataset("titanic")
sns.boxplot(x="class", y="age", hue="sex", data=titanic)
plt.legend(loc="best")
plt.show()
```

여러가지 boxplot

```python
# boxenplot

sns.boxenplot(x="sex", y="tip", hue="smoker", data=tips)
```

```python
# swarmplot

sns.swarmplot(x="day", y="tip", hue="smoker",
              alpha=0.7, data=tips)
# alpha: 투명도(transparency) 조정
```

```python
# violin plot: 카테고리컬한 데이터의 분포를 시각적으로 잘 비교할 수 있도록 만들어진 차트

sns.violinplot(x="day", y="tip", hue="smoker",
               bw=0.25, split=True, data=tips)

# bw: bandwidth(대역폭)의 줄임말. 이를 설정해줌으로써 더 잘 fit 시켜줄 수 있음
# split: split=True를 해주면 대칭성을 이용하여 합쳐준다.(이를 안써주면 파란색, 붉은색 그래프가 따로 나옴)
```

## 여러가지 boxplot

- boxenplot 
  - boxplot과 거의 같음
  - boxplot에 데이터의 분포에 따른 박스의 크기를 반영


- swarmplot
  - 실제 데이터가 몇개나 분포되어있는지 확인하기 좋다.
  - boxplot과 함께 그려보면 좋음

- violinplot
  - 카테고리컬한 데이터의 분포를 시각적으로 잘 비교할 수 있도록 만들어진 차트
